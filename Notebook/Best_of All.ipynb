```python
import pandas as pd

# Data for the models
data = {
    'Model': [
        'Random Forest with SMOTE + ENN',
        'CNN with SMOTEENN',
        'RNN with SMOTE+ENN',
        'SVM with SMOTEENN',
        'FNN with SMOTEENN',
        'XGBoost with SMOTEENN'
    ],
    'Recall': [0.72973,0.771499,0.707617,0.791155,0.769042,0.872236],
    'Precision': [0.492537,0.448571,0.337632,0.424242,0.475684,0.344995],
    'F1-Score': [0.588119,0.567299,0.457143,0.552316,0.587793,0.494429],
    'F2-Score':[0.665621,0.674399,0.580411,0.674487,0.684602,0.668047],
    'Accuracy': [0.792,0.7605,0.658,0.739,0.7805,0.637],
    'AUC-ROC': [0.850495,0.852296,0.743321,0.838472,0.859092,0.827034]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 14))

# List of metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'F2-Score', 'AUC-ROC']

# Plotting each metric in a separate subplot
for idx, metric in enumerate(metrics):
    sns.barplot(x='Model', y=metric, data=df, ax=axes[idx//2, idx%2])
    axes[idx//2, idx%2].set_title(f'{metric} Comparison')
    axes[idx//2, idx%2].set_xticklabels(axes[idx//2, idx%2].get_xticklabels(), rotation=45, ha="right")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Plotting a line plot for each metric across models
plt.figure(figsize=(12, 8))
for metric in metrics:
    plt.plot(df['Model'], df[metric], marker='o', label=metric)

plt.title('Model Performance Metrics')
plt.xlabel('Model')
plt.ylabel('Score')
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.grid(True)
plt.show()
```
